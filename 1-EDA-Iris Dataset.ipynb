# Mean,Variance and Standard Deviation

```python
#iris.Species.value_counts()
#iris_setosa=iris[Species=="Iris-setosa"]
iris_setosa=iris.loc[iris["Species"]=="Iris-setosa"]
iris_virginica=iris.loc[iris["Species"]=="Iris-virginica"]
iris_versicolor=iris.loc[iris["Species"]=="Iris-versicolor"]
```

```python
#Mean is simple average
#Mean is telling about Central Tendency
print("Means")
print(np.mean(iris_setosa["PetalLengthCm"]))


#Mean with an outlier
#We added one value 50, there is drastic increase in mean
#Problem of Mean- Because of one outlier, Mean can increase drastically
print(np.mean(np.append(iris_setosa["PetalLengthCm"],50)));
print(np.mean(iris_virginica["PetalLengthCm"]));
print(np.mean(iris_versicolor["PetalLengthCm"]))



#The range i.e spread of Setosa PetalLengthCMm is less when compared to 
# range/spread of versicolor and virginica PetalLengthCm
#the width is what we call as spread
#Spread is called AS VARIANCE 
#Setosa spread is less
#versicolor is widely spread than setosa, but less than virginica
#virginica is more widely spread

#Standard Deviation
print("\n Standard Deviation")
print(np.std(iris_setosa["PetalLengthCm"]))
print(np.std(iris_virginica["PetalLengthCm"]));
print(np.std(iris_versicolor["PetalLengthCm"]))

#Setosa has less spread
#virginica has little more spread
#versicolor has high spread
```

# Median,Percentile,Quantile,IQR,MAD

```python
print("Median")
print(np.median(iris_setosa["PetalLengthCm"]))


#Median with an outlier
print(np.median(np.append(iris_setosa["PetalLengthCm"],50)));
print(np.median(iris_virginica["PetalLengthCm"]));
print(np.median(iris_versicolor["PetalLengthCm"]))

print("\nQuantiles")
#we need all the percentiles between zero to hundred with a gap of 25
print(np.percentile(iris_setosa["PetalLengthCm"],np.arange(0,100,25)));
print(np.percentile(iris_virginica["PetalLengthCm"],np.arange(0,100,25)));
print(np.percentile(iris_versicolor["PetalLengthCm"],np.arange(0,100,25)))

print("\n90th percentile")
print(np.percentile(iris_setosa["PetalLengthCm"],90))
print(np.percentile(iris_virginica["PetalLengthCm"],90))
print(np.percentile(iris_versicolor["PetalLengthCm"],90)) 

from statsmodels import robust
print("\n Median Absolute Deviation")
print(robust.mad(iris_setosa["PetalLengthCm"]))
print(robust.mad(iris_virginica["PetalLengthCm"]))
print(robust.mad(iris_versicolor["PetalLengthCm"])) 
```

# Cumulative Distribution Function

```python
#We can visually see what percentage of versicolour flowers have a
#petal length of less than 1.6

counts,bin_edges=np.histogram(iris_setosa['PetalLengthCm'],bins=10,density=True)

pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)

#compute CDF
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

#Starts with zero at left bottom and ends at 1 on left top
#1 means 100% of setosa flowers have petal length less than equal to 1.9
#Orange CDF--- orange point says what percentage of flowers
#Blue PDF
#PDF counts how many flowers are there for each Petal Length-->histogram
#PDF- About 20% of points have petal lengths between 1.5 and 1.6
#CDF: 82% of setosa flowers have petal length less than 1.6
 
#If we have a PDF, the corresponding value of CDF is basically the area under the curve 
#of PDF till that point
#If we differentiate CDF we will get PDF
#If we do integration on PDF we will get CDF
```

```python
#Misclassification error if you use petal length only
counts,bin_edges=np.histogram(iris_setosa['PetalLengthCm'],bins=10,density=True)

pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

#versicolor
counts,bin_edges=np.histogram(iris_versicolor['PetalLengthCm'],bins=10,density=True)

pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)


#Virginica

counts,bin_edges=np.histogram(iris_virginica['PetalLengthCm'],bins=10,density=True)

pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()

#If petal length < 2, the flower is setosa
#100% of setosa points are less than 2 petal length

#Now coming to versicolor and virginica
#We put a threshold of 5
#CDF for versicolor is 0.95
#PDF for versicolor is 0.1
#if petal length >2 and  petal length <5 , it is versicolor
### 95% of versicolor flowers have petal length less than or equal to 5
### we will be correct with versicolor 95% of times
### we will be correct with versicolor 95% of times, wrong 5% times
#if petal length >2 and  petal length >5 , it is virginica
### But 10% of points(we get to know this by CDF of Virginica)
### If we use this rule, 10% of times we make a mistake here
### we will be correct 90% times and wrong 10% times
```
